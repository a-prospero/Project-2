```python
import pandas as pd
import json
import pprint
import requests
```

```python
nfl_df = pd.read_csv("nfl_data.csv")
key = '95a11625c7bc4b1ca1505834210902'
```

```python
my_list = nfl_df.columns.values.tolist()
print(my_list)
```

```python
unnecessary = ['first_downs_away', 'first_downs_home', 'third_downs_away', 'third_downs_home', 'fourth_downs_away',
               'fourth_downs_home', 'passing_yards_away', 'passing_yards_home', 'rushing_yards_away', 'rushing_yards_home',
               'total_yards_away', 'total_yards_home', 'comp_att_away', 'comp_att_home', 'sacks_away', 'sacks_home',
               'rushing_attempts_away', 'rushing_attempts_home', 'fumbles_away', 'fumbles_home', 'int_away', 'int_home',
               'turnovers_away', 'turnovers_home', 'penalties_away', 'penalties_home', 'redzone_away', 'redzone_home',
               'drives_away', 'drives_home', 'def_st_td_away', 'def_st_td_home', 'possession_away', 'possession_home'] 
nfl_df = nfl_df.drop(columns=unnecessary)
```

```python
nfl_df
```

```python
def who_won(row):
    if row['score_away'] > row['score_home']:
        return 0
    else:
        return 1
```

```python
nfl_df['hometeam win'] = nfl_df.apply(lambda row: who_won(row), axis=1)
```

```python
nfl_df = nfl_df.sort_values(['home'], ascending=True)
nfl_df
```

```python
teams = nfl_df['home'].values
teams = list(set(teams))
print(teams)
```

```python
teams.sort()
print(teams)
```

```python
nfl_data_transformed = pd.DataFrame(columns=['Teams', 'Homefield Wins', 'Homefield Losses', 'Home Win Percentage',
                                            'High Temperature', 'Low Temperature', 'Average Temperature'])
nfl_data_transformed['Teams'] = teams
nfl_data_transformed = nfl_data_transformed.sort_values(['Teams'], ascending=True)
nfl_data_transformed.set_index(['Teams'], inplace=True)
nfl_data_transformed
```

```python
team_wins = []
team_losses = []
total_games = []
for team in teams:
    wins_df = pd.DataFrame()
    wins_df = nfl_df.loc[nfl_df['home'] == team]
    wins_df = wins_df.loc[wins_df['hometeam win'] == 1]
    x = wins_df['hometeam win'].count()
    team_wins.append(x)
    print(f"{team} has {x} wins" )
    
    losses_df = pd.DataFrame()
    losses_df = nfl_df.loc[nfl_df['home'] == team]
    losses_df = losses_df.loc[losses_df['hometeam win'] == 0]
    y = losses_df['hometeam win'].count()
    team_losses.append(y)
    print(f"{team} has {y} losses" )
    
    total_df = pd.DataFrame()
    total_df = nfl_df.loc[nfl_df['home'] == team]
    z = total_df['home'].count()
    total_games.append(z)
    print(f"{z} total games \n")
```

```python
print(team_wins)
```

```python
# test for accuracy
subset_df = pd.DataFrame()
subset_df = nfl_df.loc[nfl_df['home'] == 'Vikings']
subset_df = subset_df.loc[subset_df['hometeam win'] == 1]
subset_df['hometeam win'].count()
```

```python
win_percentage = []
for i in range(len(team_wins)):
    x = ((team_wins[i]/total_games[i]) * 100).round(2)
    x = str(x) + '%'
    win_percentage.append(x)
               
print(win_percentage)
```
