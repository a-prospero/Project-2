```python
import pandas as pd
import csv,json
```

```python
NHL_wins = pd.read_csv('Database_Resources/NHL_HomeWins.csv')
NHL_wins
NHL_arenas = pd.read_csv('Database_Resources/NHL_arenas.csv')
NHL_arenas

nhl_home_wins = pd.merge(NHL_wins, NHL_arenas, how = 'left', on = 'Team')
NHL_home_wins = nhl_home_wins.set_index('Team')
NHL_home_wins
```

```python
nhl_dict = nhl_home_wins.to_dict('index')
nhl_dict
```

```python
with open("nhl_home_wins.json", "w") as outfile:
    json.dump(nhl_dict, outfile)
```
